```python
import pandas as pd
df = pd.read_csv('football_simulated.csv')
print(df.head())
print(df.info())
print(df.describe()) #having observed the weight the 50% 75% and max for weight is 82kg, 
                     #indicating that data is lost on heavier players and that the data was clipped at a max of 82kg
print(df.isna().sum()) #this line of code shows the emply cells of each column, there are a few

#Here I have identified a spelling mistake in the column and am changing it in the column header
df.rename(columns={'ThighCircumerence(cm)': 'Thigh_Circumference_cm'}, inplace=True)

#now I´m checking for duplicates, and there are none

duplicate_rows = df.duplicated() #This shows duplicates, there are none
num_duplicates = duplicate_rows.sum
print(num_duplicates) #No duplicates found


#In the following I´m checking to see if the data has missing cells in more than one row, if so I´ll simply delete them 
missing_cells_per_row = df.isnull().sum(axis=1)
rows_with_multiple_missing = missing_cells_per_row[missing_cells_per_row > 1]

indices_of_rows_with_multiple_missing = rows_with_multiple_missing.index.tolist()
if not rows_with_multiple_missing.empty:
        print(f"\nFound {len(rows_with_multiple_missing)} row(s) with more than one missing cell.")
        print("Indices of these rows and their missing cell count:")
        print(rows_with_multiple_missing)

        print("\nThese are the actual rows with more than one missing cell:")
        print(df.loc[indices_of_rows_with_multiple_missing])
        df.drop(index=indices_of_rows_with_multiple_missing, inplace=True) #This line deltes the rows with the two or more missing cells
else:
        print("\nNo rows found with more than one missing cell.")


#Now I want to populate the missing cells with the mean of that column, 
#and because there is no catgorical data I can do this easily
#however I´ll go column by column to 2 decimal places
df['Thigh_Circumference_cm'] = df['Thigh_Circumference_cm'].fillna(round(df['Thigh_Circumference_cm'].mean(), 2))
df['Speed(m/s)'] = df['Speed(m/s)'].fillna(round(df['Speed(m/s)'].mean(), 2))
df['Age'] = df['Age'].fillna(round(df['Age'].mean(), 2))


print(df.describe())

#Having checked the whole file with the code above is full, Now I´ll create a newfile to replect this

df.to_csv('football_sim_cleaned.csv', index=False) #index is false to ensure the header is not index 0

print("Cleaned data successfully saved to 'football_sim_cleaned.csv'")


```
